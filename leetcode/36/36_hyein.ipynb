```python
class Solution:
    def isValidSudoku(self, board: List[List[str]]) -> bool:
        for i in range(9):
            tmp = []
            for j in board[i]:
                if j != '.':
                    tmp.append(j)
            len_val = len(tmp)
            len_val_set = len(set(tmp))
            if len_val != len_val_set:
                return False
            
        for i in range(9):
            tmp = []
            for j in range(9):
                if board[j][i] != '.':
                    tmp.append(board[j][i])  
            cnt_row_val = len(tmp)
            cnt_row_val_set = len(set(tmp))
            if cnt_row_val != cnt_row_val_set:
                return False
            
        for i in range(0,9,3):
            for j in range(0,9,3):
                    tmp = []
                    for k in range(3):
                        for l in range(3):
                            if board[i+k][j+l] != '.':
                                tmp.append(board[i+k][j+l])         
                    cnt_3x3 = len(tmp)
                    cnt_3x3_set = len(set(tmp))
                    if cnt_3x3 != cnt_3x3_set:
                        return False
        return True
```
