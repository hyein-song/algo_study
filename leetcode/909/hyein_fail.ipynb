```python
# DP 최종 ㅠㅠ 실패

class Solution:
    def snakesAndLadders(self, board) -> int:
        N = len(board[0])
        b2 = [0]
        for i in range(N-1,-1,-1):
            if i%2 != N%2 :
                for j in range(N):
                    b2.append(board[i][j]) 
            else:
                for j in range(N-1,-1,-1):
                    b2.append(board[i][j])
        # invalid 
        dst = []
        for i in range(2, N*N):
            if b2[i] != -1:
                dst.append(b2[i])
            else:
                for j in range(1,7):
                    if i+j <= N*N :
                        dst.append(i+j)              
        if N*N not in dst:
            return -1                                     
                    
        dp_org = [0] * (N*N+1)
        dp_org[1] = 0
        dp_org[2] = 1
        
        flag = [0] * ((N*N)+1)
        
        # init
        for i in range(3,(N*N)+1):
            tmp = []
            for j in range(1,7):
                if i-j < 1 :
                    continue
                tmp.append(dp_org[i-j]+1)   
            dp_org[i] = min(tmp) 
                                  
        # snake, ladder
        for i in range(2,(N*N)+1):
            tmp = []
                    
            if b2[i] != -1:
                for j in range(1,7):
                    if i-j <= 1 :
                        continue
                    tmp.append(dp_org[i-j]+1)
                    
                if flag[i] == 0 :    
                    tmp.append(dp_org[i])
                else:
                    tmp.append(dp_org[i]+1)    

                tmp.append(dp_org[b2[i]])    
                dp_org[b2[i]] = min(tmp)
                # print(i,b2[i], dp_org[b2[i]])
                flag[b2[i]] = 1
                
                # if b2[i]==N*N:
                #     return dp_org[b2[i]]
            else:
                tmp = []
                for j in range(1,7):
                    if i-j < 1 :
                        continue
                    tmp.append(dp_org[i-j]+1)
                tmp.append(dp_org[i])     
                dp_org[i] = min(tmp)

            # print(i, dp_org)    

        return dp_org[N*N]

a = Solution()
board = [[-1,-1,27,13,-1,25,-1],
[-1,-1,-1,-1,-1,-1,-1],
[44,-1,8,-1,-1,2,-1],
[-1,30,-1,-1,-1,-1,-1],
[3,-1,20,-1,46,6,-1],
[-1,-1,-1,-1,-1,-1,29],
[-1,29,21,33,-1,-1,-1]]
print(a.snakesAndLadders(board))
```
